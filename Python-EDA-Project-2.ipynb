```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace= True)
```

```python
df.dropna(inplace=True)
```

```python
df["Amount"]=df["Amount"].astype("int64")
```

```python
ax=sns.countplot(x='Gender',data=df,hue='Gender')
```

```python
ax=sns.countplot(x='Gender',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
salesdf=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Gender',y='Amount',data=salesdf,hue='Gender')
plt.show()
```

```python
ax=sns.countplot(data=df,x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
salesdf=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Age Group',y='Amount',data=salesdf,hue="Age Group")
plt.show()
```

```python
salesdf=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='State',y='Orders',data=salesdf,hue="State")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

```python
salesdf=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='State',y='Amount',data=salesdf,hue="State")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

```python
ax=sns.countplot(x='Marital_Status',data=df,hue='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
salesdf=df.groupby(['Marital_Status'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
ax=sns.barplot(x='Marital_Status',y='Amount',data=salesdf,hue="Marital_Status")
plt.show()
```

```python
salesdf=df.groupby(['Marital_Status'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
ax=sns.barplot(x='Marital_Status',y='Amount',data=salesdf,hue="Marital_Status")
plt.show()
```

```python
salesdf=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='Occupation',y='Amount',data=salesdf,hue="Occupation")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.countplot(x='Product_Category',data=df,hue='Product_Category')
ax.set_xticklabels(ax.get_xticklabels(),rotation=80)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
salesdf=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Product_Category',y='Amount',data=salesdf,hue='Product_Category')
ax.set_xticklabels(ax.get_xticklabels(),rotation=80)
plt.show()
```

```python
salesdf=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,3)})
ax=sns.barplot(x='Product_ID',y='Orders',data=salesdf,hue='Product_ID')
ax.set_xticklabels(ax.get_xticklabels(),rotation=60)
plt.show()
```
